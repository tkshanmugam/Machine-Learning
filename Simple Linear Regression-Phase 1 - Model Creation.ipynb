```python
# import the pandas library bcoz pandas has database and table related functions
import pandas as pd
```

```python
# Here read the Salary_Data.csv by using pandas function and assigned values to the dataset variable.
dataset = pd.read_csv("Salary_Data.csv")
```

```python
dataset
```

```python
# Step - Split input and output
```

```python
#Here assigned values of YearsExperience column to the independent variable (Split input)
independent = dataset[["YearsExperience"]]
independent
```

```python
#Here assigned values of Salary column to the dependent variable (Split output)
dependent = dataset[["Salary"]]
dependent
```

```python
#Split the train and test test set
from sklearn.model_selection import train_test_split
x_train, X_test, y_train, y_test = train_test_split(independent, dependent, test_size=0.30, random_state =0)
```

```python
#create model
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train) #fit function is used to create model
```

```python
weight = regressor.coef_ # find the weight, here weight means distance between two points
weight 
```

```python
bais = regressor.intercept_ # find the bias, here bias means minium value or starting value or initial value
bais
```

```python
y_pred = regressor.predict(X_test) #find the predict value for test
```

```python
# Step - Evalution Metrics
```

```python
# load r2_score library for evaluate the metrics
from sklearn.metrics import r2_score 
# Test the created model by using r2_score library
r_score = r2_score(y_test, y_pred)
```

```python
r_score
```

```python
# Step - Save Best Model
```

```python
# load pickle library
import pickle
filename = "finalized_model_linear.sav" # assign value to the filename variable
```

```python
pickle.dump(regressor,open(filename,'wb')) # transfer created model to the file, wb - write binary
```

```python
loaded_model = pickle.load(open('finalized_model_linear.sav', 'rb')) # Read the save model for predict
result = loaded_model.predict([[13]])
```
