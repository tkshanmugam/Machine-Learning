```python
import pandas as pd
import numpy as np
```

```python
# Load the dataset (assuming it's in a CSV file)
data = pd.read_csv('../1.DataCollection/onlinefoods.csv')
```

```python
#Identify Missing Values
data.isnull().sum()
```

```python
def quanQual(dataset):
    quan=[]
    qual=[]
    for columnName in dataset.columns:
    #print(columnName)
        if(dataset[columnName].dtypes=='O'):
        #print("qual")
            qual.append(columnName)
        else:
        #print("quan")
             quan.append(columnName)
    return quan,qual
```

```python
quan,qual=quanQual(data)
```

```python
def detect_outliers_iqr(data):
    # Calculate the first quartile (Q1) and third quartile (Q3)
    Q1 = np.percentile(data, 25)
    Q3 = np.percentile(data, 75)
    # Calculate the interquartile range (IQR)
    IQR = Q3 - Q1
    # Define the lower and upper bounds for outliers detection
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    # Find outliers
    outliers = [x for x in data if x < lower_bound or x > upper_bound]
    return outliers
```

```python
for colunName in quan:
    outliers_iqr = detect_outliers_iqr(data[colunName])
    print("Outliers detected using IQR method :",colunName,  outliers_iqr)
```
